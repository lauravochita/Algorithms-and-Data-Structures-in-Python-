````python
def load_data(filepath):
    import csv
    raw_data = []
    with open(filepath, 'r', encoding="utf8") as rf:
        reader = csv.reader(rf, delimiter = ',')
        for row in reader:
            raw_data.append(row)
    header = raw_data[0]
    data = raw_data[1:]

    return header, data
````

2. If nothing matches the search request, return an empty list.

````python
def extract_subset(table, header, request):
    results = []
    for row in table:
        check = True
        for key in request:
            if key in header:
                if type(request[key]) == list:
                    if row[header.index(key)] not in request[key]:
                        check = False
                else:
                    if row[header.index(key)] != request[key]:
                            check = False
            else:
                print("Invalid request")
                return []

        if check:
         results.append(row)
    return results
````
